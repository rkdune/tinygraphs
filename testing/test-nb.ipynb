```python
import tinyGraphs as tg
```

```python
train_losses, val_losses = [], []
epochs = 3

for epoch in range(epochs):
    # Simulate training loss decreasing linearly
    train_loss = 10 - (epoch * 2)  # Starts at 10, decreases by 2 each epoch
    train_losses.append(train_loss)
    
    # Simulate validation loss with some noise
    val_loss = train_loss + 1 + (epoch * 0.5)  # Slightly higher than training, increases slowly
    val_losses.append(val_loss)
    
    # Plot using tinyGraphs
    tg.plot(train_losses, val_losses, epoch, legend_loc = "upper right", updating_title = True, 
         legend = True, x_label = "Epoch", y_label = "Loss", title = "Training Progress", dark_mode = True, color_scheme = "jet")
```
